### 5.3 Error Analysis (5 points)

Include:
- patterns in residuals or prediction errors
- overprediction or underprediction

**5.3 Answer:**  
5.3 Error Analysis

The residual plots reveal a clear pattern of heteroscedasticity, visualized as a fan shape where the spread of errors widens significantly as the predicted home value increases. This indicates that the model's precision is highly dependent on the price tier of the home. For properties under $500,000, the residuals are relatively tightly clustered. However, for homes approaching $1,000,000, the error margins expand drastically. This happens because a standard 10% prediction error on a $400,000 home is $40,000, whereas the same percentage error on a $1.5 million home is $150,000, causing the absolute variance to naturally scale with the property's magnitude.

Beyond just the widening variance, the errors are not symmetrical. There is a systematic under-prediction of expensive homes. As actual home values cross the $800,000 threshold, the model consistently guesses too low. This is a common pitfall when algorithms minimize overall error across a dataset dominated by median-priced homes. The model essentially learns to play it safe by anchoring its predictions closer to the global average. When faced with a luxury property, it lacks the statistical confidence to predict an extreme value, resulting in the invisible prediction ceiling observed around $1.3 million.

This under-prediction is closely tied to the outliers and the heavily right-skewed nature of the target variable. Rare, high-value homes are not just larger versions of average homes; they often possess fundamentally different characteristics that drive their price exponentially rather than linearly. Features like panoramic views, custom architectural builds, or highly specific neighborhood prestige are critical value drivers for the luxury market but are largely absent from this dataset. Because the model only has access to foundational features like square footage and location, it cannot mathematically justify the massive price premiums these outliers command in the real estate market.

To address these limitations, future work should take an iterative approach to refining the target variable and expanding the dataset. A practical next step would be applying a log transformation to the home prices before training, which would help normalize the rightward skew and allow the model to learn relative percentage errors rather than absolute dollar amounts. Additionally, the clear divide in error patterns suggests that a segmented modeling strategy might be necessary, training one model for the general market and a separate model specialized for properties over $800,000. Finally, bringing in external data to capture missing qualitative features, such as school district ratings or topographical view quality, would provide the algorithm with the context required to accurately price top-tier real estate.
